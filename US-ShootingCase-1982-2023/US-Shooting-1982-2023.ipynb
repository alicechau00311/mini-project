## 1.2 Data Cleaning

### 1.2.1 Change data type

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df = df.convert_dtypes()
df.info()
```

```python
df.rename(columns = {'location.1':'place'}, inplace = True)
```

```python
df.columns
```

```python
df[['place', 'prior_signs_mental_health_issues', 'weapons_obtained_legally', 'race', 'type']] = df[['place', 'prior_signs_mental_health_issues', 'weapons_obtained_legally', 'race', 'type']].apply(lambda col:col.str.capitalize())
df[['place', 'prior_signs_mental_health_issues', 'weapons_obtained_legally', 'race', 'type']].head()
```

```python
df['gender'] = df['gender'].replace('M', 'Male')
df['gender'] = df['gender'].replace('F', 'Female')
df['gender']
```

```python
df[df['gender'].map(len) > 6]['gender']
```

Index 3 value will be convert to Female for converience calculation

```python
df['gender'][3] = 'Female'
df['gender']
```

### 1.2.2 Clean Missing Data

'-' is the missing data so it will be convert to "NaN".

```python
df = df.replace('-', np.nan)
# display columns which is null
df.loc[:, df.isna().any()]
```
